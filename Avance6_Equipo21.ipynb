# **Avance 1. Análisis exploratorio de datos (EDA - Exploratory Data Analysis)**

El Análisis Exploratorio de Datos (EDA, por sus siglas en inglés) es un proceso fundamental en la ciencia de datos que permite entender las características principales de un conjunto de datos antes de realizar modelado o análisis predictivo. Su objetivo es revelar patrones, relaciones y posibles anomalías, así como generar hipótesis y preparaciones para el análisis avanzado.

### Objetivos del EDA
* Conocer el Contexto de los Datos: Entender de qué trata cada variable, sus tipos de datos, y qué representan en el contexto del problema.
* Estadísticas Descriptivas: Calcular y examinar estadísticas básicas como la media, mediana, moda, desviación estándar, y percentiles para cada variable. Esto ayuda a obtener una visión general de la distribución de los datos.
* Visualización de Datos: Usar gráficos como histogramas, boxplots, gráficos de dispersión, y mapas de calor para identificar patrones, correlaciones y distribuciones. Las visualizaciones facilitan la comprensión de la estructura de los datos y destacan posibles outliers.
* Identificación de Outliers y Datos Faltantes: Detectar valores atípicos (outliers) y datos faltantes que pueden requerir limpieza o tratamiento antes del análisis o modelado.
* Comprender Relaciones entre Variables: Evaluar la correlación y las relaciones entre variables numéricas y categóricas para identificar posibles interacciones o variables relevantes para el análisis posterior.

###Principales Herramientas y Técnicas en EDA
* Estadísticas Descriptivas: Cálculo de métricas estadísticas básicas (mean, median, std, min, max, etc.) para obtener una visión rápida de las características de cada variable.
* Visualización de Datos:
  * Histogramas para ver la distribución de datos en cada variable.
  * Boxplots para identificar outliers y rango intercuartílico.
  * Gráficos de dispersión para observar la relación entre dos variables numéricas.
  * Mapas de calor de correlación para visualizar las relaciones entre múltiples variables.
* Manejo de Datos Faltantes: Identificar y tratar los valores faltantes mediante imputación (relleno de datos faltantes) o eliminación de filas/columnas si es necesario.
* Identificación de Outliers: Detección de valores fuera de rango que pueden afectar el análisis y que a menudo requieren tratamiento especial.

###Beneficios del EDA
El EDA es crucial para crear una base sólida para el análisis o modelado de datos. Algunos beneficios incluyen:
* Mejor Comprensión de los Datos: Proporciona un entendimiento profundo de las características y peculiaridades de los datos.
* Detección Temprana de Problemas: Identifica errores, datos atípicos, y otros problemas que podrían afectar los resultados del modelo.
* Optimización de Modelos: Facilita la selección de variables y puede sugerir transformaciones o preprocesamientos que mejorarán el rendimiento de los modelos.

````python
import os
import pandas as pd

# Define the directory path where the Excel files are stored
excel_file_path = '/content/sample_data/'

# Define the expected column names to standardize the structure of the data
column_names = [
    '# Registro', 'Fecha de adquisición', 'Hora Adquisición', 'Temperatura Ambiente  (°C)',
    'Inverter fault (fault=1)', 'Inverter running', 'Brake contactor feedback (open=1/close=0)',
    'Travelling direction (fwd=0/rev=1)', 'Referencia Velocidad', 'Velocidad Actual',
    'Relative load cell actual weight [%]', 'Delta Temperatura por Velocidad (°C) Rod 1',
    'Temperatura Rod 1', 'Delta Temperatura por Velocidad (°C) Rod 2', 'Temperatura Rod 2',
    'Delta Temperatura por Velocidad (°C) Rod 3', 'Temperatura Rod 3', 'Vibracion punto 1',
    'Vibracion punto 2', 'Vibracion punto 3', 'Motor actual relative current (%)',
    'Motor actual relative current (%).1'
]

# Define a dictionary for renaming specific columns in each file
rename_dict = {
    'Motor actual relative current (%) FWD': 'Motor actual relative current (%)',
    'Motor actual relative current (%) REVERSE': 'Motor actual relative current (%).1'
}

# List to store dataframes created from each Excel file
all_dataframes = []

# Iterate through each file in the specified directory
for filename in os.listdir(excel_file_path):
    if filename.endswith('.xlsx'):
        file_path = os.path.join(excel_file_path, filename)
        try:
            # Read the Excel file into a pandas DataFrame
            df = pd.read_excel(file_path)

            # Rename columns as per the rename_dict to ensure consistency
            df.rename(columns=rename_dict, inplace=True)

            # Add any missing columns from column_names to ensure uniform structure
            for column in column_names:
                if column not in df.columns:
                    df[column] = None  # Fill missing columns with None

            # Reorder columns to match column_names
            df = df[column_names]

            # Append the processed DataFrame to the list
            all_dataframes.append(df)
        except Exception as e:
            print(f"Error reading file {filename}: {e}")

# Concatenate all DataFrames if any were successfully processed
if all_dataframes:
    merged_df = pd.concat(all_dataframes, ignore_index=True)
    print("Files concatenated successfully.")
else:
    print("No matching files with expected columns found in the specified path.")

# Display the first few rows of the merged DataFrame
merged_df.head()
````

En las siguientes funciones vamos a determinar si tenemos datos nulos, contantes y outliers.

````python
import pandas as pd

def check_missing_data(df):
    # Calculate the total number of missing (null) values in each column
    missing_data = df.isnull().sum()

    # Calculate the percentage of missing values for each column
    missing_percentage = (missing_data / len(df)) * 100

    # Create a summary DataFrame with total and percentage of missing values per column
    missing_summary = pd.DataFrame({
        'Total Missing': missing_data,
        'Percentage Missing': missing_percentage
    })

    # Filter the summary to include only columns with missing values
    missing_summary = missing_summary[missing_summary['Total Missing'] > 0]

    # Return the summary sorted by percentage of missing data in descending order
    return missing_summary.sort_values(by='Percentage Missing', ascending=False)

# Usage
missing_summary = check_missing_data(df)
print("Missing Data by Column:\n", missing_summary)

````

````python
def check_constant_columns(df):
    # Identify columns where all values are the same (constant columns)
    constant_columns = [col for col in df.columns if df[col].nunique() == 1]

    # Return the list of constant columns
    return constant_columns

# Usage
constant_columns = check_constant_columns(df)
print("Columns with constant values:\n", constant_columns)

````

````python
def detect_outliers(df, threshold=1.5):
    # Initialize a dictionary to store outlier counts for each column
    outliers = {}

    # Loop through numeric columns (integer and float types only)
    for col in df.select_dtypes(include=['float', 'int']).columns:
        # Calculate the first (Q1) and third quartiles (Q3) and the interquartile range (IQR)
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        # Determine lower and upper bounds for detecting outliers
        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR

        # Count the number of outliers that fall below the lower bound or above the upper bound
        outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col].count()

        # Add to outliers dictionary if any outliers are detected in this column
        if outliers_count > 0:
            outliers[col] = outliers_count

    # Create a summary DataFrame of outliers counts per column
    outliers_summary = pd.DataFrame.from_dict(outliers, orient='index', columns=['Outliers Count'])
    return outliers_summary

# Usage
outliers_summary = detect_outliers(df)
print("Outliers Summary:\n", outliers_summary)

````

````python
# Define limits for each column in a dictionary
limits = {
    'Referencia Velocidad': {'lower': 0, 'upper': 100},
    'Velocidad Actual': {'lower': 0, 'upper': 100},
    'Relative load cell actual weight [%]': {'lower': 0, 'upper': 100},
    'Temperatura Rod 1': {'lower': -10, 'upper': 110},
    'Temperatura Rod 2': {'lower': -10, 'upper': 110},
    'Temperatura Rod 3': {'lower': -10, 'upper': 110},
    'Vibracion punto 1': {'lower': 0, 'upper': 0.15},
    'Vibracion punto 2': {'lower': 0, 'upper': 0.15},
    'Vibracion punto 3': {'lower': 0, 'upper': 0.15}
}

# Function to check if any values exceed specified limits and add a target column 'Y'
def check_limits(df, limits):
    # Initialize a new column 'Y' with 0, assuming all values are within limits
    df['Y'] = 0

    for column, bounds in limits.items():
        lower = bounds.get('lower')
        upper = bounds.get('upper')

        # Update 'Y' to 1 if any value in the column exceeds the lower or upper limit
        df['Y'] = df['Y'] | df[column].apply(lambda x: 1 if (x < lower or x > upper) else 0)

    return df

# Apply the function to the original DataFrame
df = check_limits(final_df, limits)

# Display the first few rows of the modified DataFrame with the 'Y' column
df.head()
````

# **Avance 2. Ingeniería de características (FE - Feature Engineering)**

El Feature Engineering (Ingeniería de Características) es el proceso de transformar, seleccionar y crear nuevas variables (features o características) a partir de los datos originales con el objetivo de mejorar el rendimiento de los modelos de machine learning. Este paso es crucial en el desarrollo de modelos predictivos, ya que las características bien diseñadas y relevantes pueden tener un impacto significativo en la precisión y generalización de un modelo.

###Objetivos del Feature Engineering
* Extraer Información Relevante: Convertir datos en bruto en características que mejor representen los patrones del problema.
* Simplificar la Representación del Problema: Crear características que hagan el modelo más simple o más fácil de interpretar.
* Mejorar la Calidad del Modelo: Generar características que puedan ayudar al modelo a capturar relaciones complejas y mejorar su capacidad predictiva.

###Principales Técnicas de Feature Engineering

**Transformación de Datos:**

* Estandarización y Normalización: Escalar los datos para que tengan una media y una desviación estándar uniformes o para que estén en un rango específico. Esto es útil en modelos sensibles a la escala, como redes neuronales y modelos de distancia.
* Transformaciones logarítmicas y polinómicas: Aplicar transformaciones como logaritmos o raíces cuadradas para variables con distribuciones sesgadas o crear nuevas características cuadráticas/cúbicas que capturen relaciones no lineales.

**Manejo de Datos Faltantes:**

* Imputación: Rellenar valores faltantes con la media, mediana o moda de la columna, o incluso con valores específicos (como un indicador de "desconocido").
* Indicadores de datos faltantes: Crear columnas binarias que indiquen si un valor estaba o no presente, lo cual puede ser informativo.

**Codificación de Variables Categóricas:**

* One-Hot Encoding: Convertir variables categóricas en variables binarias (por ejemplo, "rojo", "azul", "verde" se convierten en columnas separadas).
* Target Encoding: Reemplazar valores categóricos con el promedio de la variable objetivo (especialmente en problemas de clasificación o regresión).

**Extracción de Características Temporales:**

Para datos con información de tiempo, se pueden extraer características como el día, la semana, el mes, la estación, si es fin de semana o incluso transformaciones como "hora del día". Estas pueden ser útiles en análisis de series temporales.

**Interacciones entre Variables:**

Crear características que representen la interacción entre dos o más variables puede ayudar al modelo a capturar relaciones complejas. Por ejemplo, en vez de usar solo el ingreso y la edad de una persona, podrías agregar una característica que sea la multiplicación o el cociente entre ambas.

**Reducción de Dimensionalidad:**

* PCA (Análisis de Componentes Principales): Reduce la cantidad de características en un conjunto de datos mientras se mantiene la mayor parte de la información.
* Selección de Características (Feature Selection): Eliminar características irrelevantes o redundantes, lo cual es útil para evitar el sobreajuste y mejorar la eficiencia del modelo.

**UMBRAL DE VARIANZA**

El umbral de varianza es una técnica de selección de características en el preprocesamiento de datos que ayuda a identificar y eliminar aquellas variables que tienen poca o nula variabilidad en sus valores. Su propósito es reducir la dimensionalidad de los datos al descartar columnas que no contribuyen significativamente al análisis, ya que las variables con baja varianza suelen contener poca información útil y no ayudan a distinguir entre distintas clases o patrones en los datos.

**Objetivo del Umbral de Varianza**

El objetivo principal de aplicar un umbral de varianza es optimizar el conjunto de datos para que sea más eficiente y preciso en las etapas de análisis y modelado. Las columnas con baja varianza tienden a ser redundantes o constantes, lo cual puede añadir ruido o aumentar la complejidad del modelo sin mejorar su capacidad de generalización. Al establecer un umbral mínimo de variabilidad, solo se mantienen las características que contienen una cantidad significativa de información.

**Resultados Esperados**

Al aplicar el umbral de varianza en un conjunto de datos, se esperan los siguientes resultados:

* Reducción de Dimensionalidad: Eliminación de columnas constantes o con baja variabilidad, lo cual reduce el número de características y simplifica el conjunto de datos.

* Mejora en la Eficiencia Computacional: Un conjunto de datos con menos columnas es más rápido de procesar y permite que los algoritmos de aprendizaje automático entrenen y ejecuten predicciones de manera más eficiente.

* Aumento en la Precisión del Modelo: Al eliminar características no informativas, se reduce el ruido en los datos, lo cual puede llevar a una mejora en el rendimiento y la precisión de los modelos predictivos.

* Facilidad de Interpretación: Con menos variables, el análisis se vuelve más manejable y comprensible, permitiendo concentrarse en las características más relevantes para el problema de estudio.

````python
from sklearn.feature_selection import VarianceThreshold

# Define the variance threshold
threshold = 0.01  # This is an example; adjust as needed

# Create a feature selector based on the variance threshold
selector = VarianceThreshold(threshold)

# Apply the selector to the DataFrame (excluding the 'Y' column)
X = df.drop(columns=['Y'])
X_selected = selector.fit_transform(X)

# Get the names of columns with variance above the threshold
selected_features = X.columns[selector.get_support(indices=True)]

# Create a new DataFrame with only the selected columns
df_selected = df[selected_features]

# Display the selected columns based on the variance threshold
print("Selected columns based on the variance threshold:")
print(selected_features)

````

**Matriz de Correlación**

Una matriz de correlación es una herramienta estadística que permite medir y visualizar la relación lineal entre diferentes variables en un conjunto de datos. Cada elemento en la matriz representa el coeficiente de correlación entre un par de variables, con valores que oscilan entre -1 y 1:

Valores cercanos a 1 indican una fuerte correlación positiva, lo que significa que, a medida que una variable aumenta, la otra también tiende a aumentar.
Valores cercanos a -1 representan una fuerte correlación negativa, indicando que cuando una variable aumenta, la otra tiende a disminuir.
Valores cercanos a 0 sugieren que no hay una relación lineal significativa entre las dos variables.


**Objetivo de la Matriz de Correlación**

El objetivo principal de una matriz de correlación es identificar relaciones entre variables, lo cual es esencial para el análisis exploratorio de datos y la selección de características. Esta matriz ayuda a encontrar:

* Patrones y Relaciones Lineales: Permite identificar variables que están altamente correlacionadas, ya sea de manera positiva o negativa.
* Redundancia: Al encontrar pares de variables con alta correlación, se pueden eliminar aquellas que son redundantes, reduciendo así la complejidad del modelo.
* Características Relevantes: Ayuda a seleccionar las variables más informativas para el análisis o modelado, mejorando la eficiencia y precisión de los modelos.

**Resultados Esperados**

Al interpretar una matriz de correlación, los analistas pueden:

* Identificar variables redundantes que pueden eliminarse para simplificar el modelo.
* Seleccionar variables que estén fuertemente correlacionadas con la variable objetivo para mejorar el rendimiento del modelo.
* Detectar relaciones inesperadas que podrían requerir un análisis más profundo.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create a DataFrame with only the columns selected by the variance threshold
df_var_filtered = df[df_selected.columns]

# Calculate the correlation matrix for the selected columns
correlation_matrix = df_var_filtered.corr().abs()

# Visualize the correlation matrix with a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Selected Columns")
plt.show()

# Identify and remove columns with high correlation (>0.9)
upper_triangle = correlation_matrix.where(
    np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
)

# Find columns with high correlation
high_corr_features = [column for column in upper_triangle.columns if any(upper_triangle[column] > 0.9)]

# Drop highly correlated columns from the DataFrame
df_reduced = df_var_filtered.drop(columns=high_corr_features)

# Display the columns removed due to high correlation
print("Columns removed due to high correlation:")
print(high_corr_features)

# Final result after applying variance threshold and removing high correlation columns
print("Final columns for analysis after removing high correlation:")
print(df_reduced.columns)

````

El resultado de la Matriz nos da como resultado eliminiar muchas de las variables que se ocuparon para calcular la variable objetivo, por esta razon solo para este ejemplo vamos a continuar con la propuesta pero es imporantante mencionar que solo se ocupara para este proceso para en los modelos se ocuparan todas las varibles objetivo.

**PCA**

El Análisis de Componentes Principales (PCA, por sus siglas en inglés) es una técnica de reducción de dimensionalidad utilizada en análisis de datos y aprendizaje automático. PCA transforma un conjunto de variables posiblemente correlacionadas en un conjunto más pequeño de variables no correlacionadas llamadas componentes principales. Estas componentes principales capturan la mayor parte de la variabilidad en los datos originales, permitiendo conservar la información relevante mientras se reduce la complejidad del conjunto de datos.

El método se basa en transformar los datos en una nueva base de coordenadas en la que los primeros ejes (componentes) explican la mayor parte de la variación en los datos. Esto se logra mediante operaciones matemáticas como la descomposición en valores propios de la matriz de covarianza de los datos.

**Objetivos del PCA**

El principal objetivo de PCA es simplificar un conjunto de datos sin perder información relevante. Sus objetivos específicos son:

* Reducción de Dimensionalidad: Al transformar los datos a un espacio de menor dimensión, PCA ayuda a reducir la cantidad de características, lo cual facilita el procesamiento y análisis, especialmente cuando el conjunto de datos es amplio.

* Eliminación de Redundancia: PCA reduce la colinealidad entre variables al encontrar combinaciones lineales de las características originales, eliminando así la redundancia.

* Mejora de la Interpretabilidad: PCA facilita el análisis exploratorio al permitir que los datos se representen visualmente en menos dimensiones, como en gráficos bidimensionales o tridimensionales que son más fáciles de interpretar.

* Optimización en el Desempeño de Modelos: La reducción de dimensionalidad con PCA puede mejorar la eficiencia y, en algunos casos, la precisión de los modelos de aprendizaje automático al enfocarse solo en las características que aportan variabilidad significativa.

* Reducción del Ruido: PCA puede filtrar el ruido de los datos, ya que las componentes principales suelen captar la variación significativa, dejando de lado las variaciones que podrían ser ruido.

````python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Scale the data before applying PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_reduced.drop(columns=['# Registro']))  # Exclude '# Registro' if it's only an identifier

# Define PCA to retain 95% of the variance
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

# View the amount of variance explained by each principal component
print("Variance explained by each principal component:")
print(pca.explained_variance_ratio_)

# Create a DataFrame with the principal components
pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])

# Display the principal components
print("Principal Components:")
print(pca_df.head())

````

Graficación de los componentes principales PC1 y PC2

````python
import matplotlib.pyplot as plt

# Plot the principal components
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c='blue', alpha=0.5)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Scatter Plot of PC1 vs PC2')
plt.show()
````

# **Avance 3. Baseline**

El baseline o línea base es una referencia inicial en el desarrollo de modelos de aprendizaje automático. Se utiliza como punto de partida para evaluar el rendimiento de cualquier modelo posterior y ayuda a determinar si un modelo más complejo proporciona una mejora significativa. La línea base suele ser un modelo simple o una métrica sencilla que se obtiene de los datos, y su propósito principal es establecer un estándar mínimo de precisión o error que cualquier otro modelo debería superar.

Un baseline puede construirse de varias maneras, dependiendo del tipo de problema:

* Para problemas de clasificación: un baseline común es un modelo que siempre predice la clase mayoritaria (clase más frecuente) o distribuye las clases aleatoriamente.
* Para problemas de regresión: un baseline típico es el promedio de los valores objetivo o la mediana, que sirve para evaluar qué tan bien predice un modelo en comparación con esta medida simple.

###Objetivos del Baseline
Los objetivos principales de un baseline son:

* Proveer una Comparativa Inicial: Establece un estándar mínimo que permite saber si un modelo más complejo tiene un desempeño aceptable y si aporta valor en relación con una solución básica.

* Evaluar el Proceso de Mejora: A medida que se desarrollan y ajustan modelos, el baseline sirve como referencia para ver cuánto mejora o empeora el rendimiento.

* Detectar Problemas en los Datos: Un baseline que obtiene un rendimiento similar a modelos complejos puede indicar problemas en los datos, como ruido o falta de patrones, lo cual puede requerir ajustes en el proceso de recolección o preparación de datos.

**Support Vector Classifier (SVC)**

El Support Vector Classifier (SVC) es un tipo de modelo de aprendizaje supervisado basado en máquinas de soporte vectorial (Support Vector Machines, SVM) que se utiliza para tareas de clasificación. SVC busca encontrar un hiperplano óptimo que separe las distintas clases en un conjunto de datos, maximizando la distancia (o margen) entre las clases. Este margen actúa como un espacio protector que ayuda a mejorar la precisión del modelo y su capacidad de generalizar a nuevos datos.

SVC es especialmente útil cuando los datos no son linealmente separables, ya que puede aplicar el truco del kernel, una técnica que permite proyectar los datos en un espacio de mayor dimensionalidad donde las clases pueden separarse más fácilmente. Existen diferentes tipos de kernels (lineal, polinomial, RBF, entre otros) que SVC puede utilizar para adaptar su funcionamiento a los patrones específicos de los datos.

**Objetivos de SVC**

Los principales objetivos del SVC son:

* Maximizar la Separación entre Clases: Al encontrar el margen óptimo entre clases, SVC mejora la clasificación y ayuda a reducir errores en la predicción.

* Flexibilidad en Casos no Lineales: Gracias a los distintos kernels, SVC puede modelar relaciones complejas en los datos, permitiendo una clasificación efectiva incluso en conjuntos de datos donde las clases no son linealmente separables.

* Generalización y Robustez: SVC es conocido por su capacidad de generalizar bien en datos no vistos, ya que el enfoque de maximizar el margen ayuda a evitar el sobreajuste.

**Aplicaciones de SVC**

El SVC se aplica en una variedad de áreas como reconocimiento de patrones, clasificación de imágenes, bioinformática y detección de fraudes, entre otros. Su capacidad para trabajar con datos de alta dimensionalidad y su robustez en tareas de clasificación complejas lo hacen una opción preferida para problemas de clasificación con datos estructurados y desbalanceados.

````python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split

# Separate the features and target variable
# Define 'X' as the feature set with selected columns that could impact the target outcome
# 'y' is the target variable 'Y', assumed to indicate a failure condition
X = df[['Temperatura Ambiente  (°C)', 'Referencia Velocidad',
        'Relative load cell actual weight [%]', 'Temperatura Rod 1',
        'Temperatura Rod 2', 'Vibracion punto 1', 'Vibracion punto 2']]
y = df['Y']  # Assuming 'Y' is the target variable

# Split data into training and validation sets
# Use a 70-30 split with a fixed random state for reproducibility
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

# Create the SVM model with an RBF kernel
# The RBF (Radial Basis Function) kernel allows the model to handle non-linear relationships
# 'C' is a regularization parameter (higher values can reduce margin size but may increase overfitting risk)
# 'gamma' controls the influence of each training example (using 'scale' adapts gamma to data distribution)
model = SVC(kernel='rbf', C=1, gamma='scale', random_state=42)  # Adjust 'C' and 'gamma' as needed
model.fit(X_train, y_train)  # Train the SVM model on the training data
````

````python
# Make predictions on the training and validation sets
y_pred_train = model.predict(X_train)
y_pred_val = model.predict(X_val)

# Evaluate the model performance on the training set
print("Training Performance:")
print("Accuracy:", accuracy_score(y_train, y_pred_train))  # Proportion of correctly classified instances
print("Precision:", precision_score(y_train, y_pred_train))  # Precision for positive class
print("Recall:", recall_score(y_train, y_pred_train))  # Recall for positive class
print("F1 Score:", f1_score(y_train, y_pred_train))  # Harmonic mean of precision and recall

# Evaluate the model performance on the validation set
print("\nValidation Performance:")
print("Accuracy:", accuracy_score(y_val, y_pred_val))  # Proportion of correctly classified instances
print("Precision:", precision_score(y_val, y_pred_val))  # Precision for positive class
print("Recall:", recall_score(y_val, y_pred_val))  # Recall for positive class
print("F1 Score:", f1_score(y_val, y_pred_val))  # Harmonic mean of precision and recall

````

**Decision Tree Classifier**

Es un algoritmo de aprendizaje supervisado que se utiliza para resolver problemas de clasificación y regresión. Como su nombre sugiere, un árbol de decisión clasifica o predice observaciones al dividir repetidamente los datos en subconjuntos basados en valores de características, formando una estructura similar a un árbol con nodos de decisión y ramas que culminan en nodos hoja (las decisiones finales). Este enfoque se basa en la selección de características o "atributos" que mejor dividen el conjunto de datos en categorías puras, maximizando la separación entre clases en cada etapa.

El algoritmo busca iterativamente la característica que proporciona la mayor ganancia de información o la menor impureza (medida, por ejemplo, mediante el índice Gini o la entropía), creando así divisiones óptimas en cada nodo. La sencillez y claridad de un árbol de decisión lo convierten en una herramienta popular, fácil de interpretar y útil en aplicaciones donde la transparencia del modelo es crítica.

**Objetivos del Decision Tree Classifier**

Los principales objetivos del Decision Tree Classifier incluyen:

* Maximizar la Separación entre Clases: Mediante divisiones sucesivas, el árbol busca agrupar observaciones similares y maximizar la pureza de cada grupo, generando predicciones claras.

* Interpretabilidad y Transparencia: A diferencia de modelos complejos, un árbol de decisión es fácil de interpretar y visualizar, lo que permite comprender el proceso de clasificación y analizar cómo se toman las decisiones.

* Versatilidad: Funciona bien con datos categóricos y numéricos, y se adapta a una variedad de aplicaciones, desde la detección de fraudes hasta la clasificación de clientes o diagnósticos médicos.

**Gradient Boosting Classifier**

Es un algoritmo de aprendizaje supervisado que utiliza el enfoque de boosting para crear un modelo predictivo sólido y preciso. Este modelo se basa en el ensamblaje de varios árboles de decisión "débiles" (árboles con bajo rendimiento individual), entrenados de forma secuencial. Cada árbol intenta corregir los errores cometidos por los árboles anteriores, utilizando la técnica del gradient boosting, que ajusta los árboles basándose en la dirección de mayor mejora en el error.

En cada iteración, el modelo ajusta un nuevo árbol a los errores residuales de los árboles previos, utilizando gradientes para minimizar la función de pérdida. Este proceso iterativo permite que el modelo enfoque su atención en los ejemplos difíciles de clasificar, mejorando así su rendimiento en problemas complejos y reduciendo el sesgo del modelo.

Objetivos del Gradient Boosting Classifier

* Optimizar la Precisión: Al ajustar gradualmente múltiples árboles para minimizar el error, el modelo logra una alta precisión, capturando patrones complejos en los datos.

* Reducir el Sesgo y el Error: Cada árbol nuevo ayuda a reducir el sesgo y la varianza del modelo al enfocarse en los errores de sus predecesores, mejorando así el rendimiento general.

* Flexibilidad para Problemas no Lineales: Debido a su capacidad para combinar múltiples árboles, el Gradient Boosting es adecuado para problemas con relaciones no lineales complejas.

````python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split

# Separate features and target variable
# Define 'X' as the feature set with selected columns potentially related to the target outcome
# 'y' is the target variable 'Y', assumed to indicate a failure condition
X = df[['Temperatura Ambiente  (°C)', 'Referencia Velocidad',
        'Relative load cell actual weight [%]', 'Temperatura Rod 1',
        'Temperatura Rod 2', 'Vibracion punto 1', 'Vibracion punto 2']]
y = df['Y']  # Assuming 'Y' is the target variable

# Split data into training and validation sets
# Use a 70-30 split with a fixed random state for reproducibility
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

# Create the Gradient Boosting model
# 'n_estimators' is the number of boosting stages (trees) added sequentially
# 'learning_rate' controls the contribution of each tree to the final prediction
# 'max_depth' limits tree depth to control complexity and avoid overfitting
gbm_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gbm_model.fit(X_train, y_train)  # Train the model on the training data

# Make predictions on the training and validation sets
y_pred_train = gbm_model.predict(X_train)
y_pred_val = gbm_model.predict(X_val)

# Evaluate the model performance on the training set
print("Gradient Boosting - Training Performance:")
print("Accuracy:", accuracy_score(y_train, y_pred_train))  # Proportion of correct predictions
print("Precision:", precision_score(y_train, y_pred_train))  # Precision for positive class
print("Recall:", recall_score(y_train, y_pred_train))  # Recall for positive class
print("F1 Score:", f1_score(y_train, y_pred_train))  # Harmonic mean of precision and recall

# Evaluate the model performance on the validation set
print("\nGradient Boosting - Validation Performance:")
print("Accuracy:", accuracy_score(y_val, y_pred_val))  # Proportion of correct predictions
print("Precision:", precision_score(y_val, y_pred_val))  # Precision for positive class
print("Recall:", recall_score(y_val, y_pred_val))  # Recall for positive class
print("F1 Score:", f1_score(y_val, y_pred_val))  # Harmonic mean of precision and recall
````

**Isolation Forest**

Es un algoritmo de aprendizaje automático no supervisado diseñado para la detección de anomalías, o outliers, en conjuntos de datos. A diferencia de otros métodos que buscan patrones de densidad o distribución de datos, Isolation Forest se basa en el principio de aislamiento de observaciones. Su enfoque es identificar instancias que se aíslen fácilmente del resto de los datos, un comportamiento común de las anomalías, ya que estas suelen tener características distintas o inusuales en comparación con la mayoría.

Isolation Forest construye varios árboles de decisión, conocidos como árboles de aislamiento, en los cuales los datos se dividen de manera aleatoria. Las observaciones que requieren pocas divisiones para ser aisladas se consideran anomalías, mientras que aquellas que necesitan más divisiones se consideran normales. La eficiencia de este proceso, tanto en tiempo de ejecución como en capacidad de manejo de grandes volúmenes de datos, hace que Isolation Forest sea una opción popular para la detección de anomalías en aplicaciones de alto rendimiento.

Objetivos de Isolation Forest

* Detectar Anomalías en Grandes Conjuntos de Datos: Isolation Forest está diseñado para detectar instancias atípicas en conjuntos de datos grandes, como transacciones bancarias, registros de red o datos industriales.

* Optimizar el Aislamiento de Outliers: Al construir árboles de aislamiento de forma aleatoria, el modelo busca maximizar el aislamiento de observaciones fuera de lo común, identificando anomalías con precisión y eficiencia.

* Minimizar el Costo Computacional: Isolation Forest es menos intensivo en cálculos que otros algoritmos de detección de anomalías, lo que lo hace adecuado para grandes volúmenes de datos y entornos de procesamiento en tiempo real.

````python
from sklearn.ensemble import IsolationForest
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Select relevant features for anomaly detection
X = df[['Temperatura Ambiente  (°C)', 'Referencia Velocidad',
        'Relative load cell actual weight [%]', 'Temperatura Rod 1',
        'Temperatura Rod 2', 'Vibracion punto 1', 'Vibracion punto 2']]

# Configure the Isolation Forest model
# 'contamination' defines the expected proportion of anomalies in the data
iso_forest = IsolationForest(n_estimators=100, contamination=0.1, random_state=42)
iso_forest.fit(X)  # Train the model on the selected features

# Predict anomalies (-1 indicates anomaly, 1 indicates normal)
anomaly_predictions = iso_forest.predict(X)

# Convert predictions (-1 to 1) into a binary label for comparison with 'Y'
# 1 represents an anomaly (failure), and 0 represents normal
anomaly_labels = np.where(anomaly_predictions == -1, 1, 0)  # 1 = anomaly, 0 = normal

# Evaluate the model performance based on the target variable 'Y'
print("Isolation Forest - Evaluation on Entire Dataset:")
print("Accuracy:", accuracy_score(df['Y'], anomaly_labels))  # Proportion of correct predictions
print("Precision:", precision_score(df['Y'], anomaly_labels))  # Precision for identifying anomalies
print("Recall:", recall_score(df['Y'], anomaly_labels))  # Recall for identifying anomalies
print("F1 Score:", f1_score(df['Y'], anomaly_labels))  # Harmonic mean of precision and recall
````

**K-Means**

Es un algoritmo de agrupamiento (clustering) no supervisado que se utiliza para dividir un conjunto de datos en grupos, o "clusters", basándose en la similitud de los datos. El objetivo de K-Means es organizar los datos en un número predefinido de grupos (denotado como K) de forma que los puntos dentro de cada grupo sean lo más similares entre sí y diferentes de los puntos en otros grupos. Este algoritmo es popular en el análisis exploratorio de datos para descubrir patrones, reducir la dimensionalidad o preparar los datos para otros modelos.

El proceso de K-Means comienza eligiendo aleatoriamente K centros iniciales para los grupos, llamados centroides. Luego, cada punto de datos se asigna al centroide más cercano, formando así los grupos iniciales. A continuación, se actualizan los centroides para que representen el centro medio de los puntos en cada grupo. Este proceso de asignación y actualización se repite hasta que los centroides se estabilizan o hasta que se alcanza un número máximo de iteraciones.

Objetivos de K-Means

* Agrupar Datos Similares: El objetivo principal es dividir los datos en clusters significativos donde los puntos dentro de cada grupo sean más similares entre sí que a los puntos en otros grupos.

* Minimizar la Distancia Intra-cluster: K-Means busca minimizar la variación dentro de cada grupo al reducir la distancia de cada punto a su centroide correspondiente, optimizando así la cohesión de los grupos.

* Simplificar el Análisis de Datos: Al reducir los datos en grupos, K-Means facilita la interpretación de patrones subyacentes y ayuda a identificar tendencias generales en conjuntos de datos complejos.

````python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

# Scale the data for efficient K-Means clustering
# Scaling ensures that all features contribute equally to the distance calculations in K-Means
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['Temperatura Ambiente  (°C)', 'Referencia Velocidad',
        'Relative load cell actual weight [%]', 'Temperatura Rod 1',
        'Temperatura Rod 2', 'Vibracion punto 1', 'Vibracion punto 2']])

# Define and fit the K-Means model
# 'n_clusters=2' assumes 2 clusters: one for normal data and one for potential failures
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_scaled)  # Train the K-Means model on the scaled data

# Assign each data point to a cluster
labels = kmeans.labels_

# Calculate the distance of each point to the centroid of its assigned cluster
# This distance can help identify points close or far from the center, potentially highlighting anomalies
distances = kmeans.transform(X_scaled).min(axis=1)

````

````python
# Define a threshold based on the 95th percentile of the distances
# Points with distances above this threshold will be considered anomalies
threshold = np.percentile(distances, 95)

# Identify anomalous points (potential failures)
# 'anomalies' is a boolean array where 'True' indicates an anomaly
anomalies = distances > threshold
````

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Convert anomalies to binary labels (1 for anomalies/failures, 0 for normal points)
anomaly_labels = anomalies.astype(int)

# Evaluate the performance by comparing with the target variable 'Y'
print("K-Means - Failure Detection Evaluation:")
print("Accuracy:", accuracy_score(df['Y'], anomaly_labels))  # Proportion of correct predictions
print("Precision:", precision_score(df['Y'], anomaly_labels))  # Precision for detecting failures
print("Recall:", recall_score(df['Y'], anomaly_labels))  # Recall for detecting failures
print("F1 Score:", f1_score(df['Y'], anomaly_labels))  # Harmonic mean of precision and recall
````

Este tipo de red neuronal consiste en una arquitectura básica de capas de neuronas que procesan la información de manera secuencial para aprender patrones a partir de los datos. Las redes neuronales simples son la base de modelos de aprendizaje profundo más complejos y se utilizan en una amplia variedad de tareas, desde clasificación y regresión hasta reconocimiento de patrones.

Una red neuronal simple típicamente tiene tres componentes principales:

* Capa Oculta: Procesa y transforma la información mediante pesos y funciones de activación, permitiendo que la red aprenda patrones complejos.
* Capa de Salida: Genera la predicción final en función del tipo de problema (clasificación o regresión).
Durante el entrenamiento, la red ajusta los pesos mediante un proceso de retropropagación, minimizando el error entre la predicción y el valor real. Este proceso optimiza los pesos a través de múltiples iteraciones, lo que permite que la red aprenda y generalice los patrones en los datos.

Objetivos de la Red Neuronal Simple

* Aprender y Generalizar Patrones: La red ajusta sus pesos para captar patrones en los datos, de modo que pueda hacer predicciones precisas en nuevos datos.

* Optimizar la Precisión del Modelo: Con suficiente entrenamiento y ajuste de hiperparámetros, las redes neuronales pueden proporcionar alta precisión en tareas complejas.

* Crear una Base para Modelos Más Complejos: Las redes neuronales simples son la base de arquitecturas de redes profundas, que resuelven problemas complejos al añadir capas y neuronas.

````python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Separate features and target variable
# Define 'X' as the feature set with selected columns potentially related to the target outcome
# 'y' is the target variable 'Y', assumed to indicate a binary classification target
X = df[['Temperatura Ambiente  (°C)', 'Referencia Velocidad',
        'Relative load cell actual weight [%]', 'Temperatura Rod 1',
        'Temperatura Rod 2', 'Vibracion punto 1', 'Vibracion punto 2']]
y = df['Y']  # Assuming 'Y' is the target variable

# Split data into training and validation sets
# Use a 70-30 split with a fixed random state for reproducibility
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

# Create the neural network model
# 'Sequential' allows layers to be added in sequence for a simple feed-forward network
model = Sequential()
model.add(Dense(32, input_dim=X_train.shape[1], activation='relu'))  # First hidden layer with 32 neurons and ReLU activation
model.add(Dense(16, activation='relu'))  # Second hidden layer with 16 neurons and ReLU activation
model.add(Dense(1, activation='sigmoid'))  # Output layer with 1 neuron and sigmoid activation for binary classification

# Compile the model
# 'adam' is an adaptive learning rate optimizer; 'binary_crossentropy' is the loss function for binary classification
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
# 'epochs=50' specifies 50 training iterations over the entire dataset
# 'batch_size=10' indicates the number of samples per gradient update
# 'validation_data' is used to evaluate the model on unseen data after each epoch
history = model.fit(X_train, y_train, epochs=50, batch_size=10, validation_data=(X_val, y_val), verbose=1)

````

## **Modelo de Emsamble**
Un modelo de ensamble es una técnica de aprendizaje automático que combina múltiples modelos base para crear un modelo más robusto y preciso. La idea fundamental detrás de los modelos de ensamble es que al combinar modelos individuales, sus errores específicos tienden a cancelarse entre sí, produciendo una predicción final que supera a los modelos individuales en términos de precisión y estabilidad. Este enfoque se ha convertido en una herramienta popular para resolver problemas complejos en diversas áreas, como la clasificación, la regresión y la detección de anomalías.

###Existen principalmente dos tipos de métodos de ensamble:

**Bagging (Bootstrap Aggregating):**

En este método, cada modelo en el ensamble se entrena en diferentes subconjuntos del conjunto de datos (generalmente mediante muestreo con reemplazo).
Los resultados de todos los modelos se combinan (por promedio en regresión o votación en clasificación) para producir la predicción final.
Ejemplo común: Random Forest, que utiliza múltiples árboles de decisión para mejorar la estabilidad y reducir el sobreajuste.

**Boosting:**

Boosting entrena modelos de forma secuencial, donde cada modelo intenta corregir los errores de los modelos anteriores.
Los modelos en Boosting suelen tener más peso en aquellos ejemplos difíciles de clasificar.
Ejemplo común: Gradient Boosting y AdaBoost, que ajustan varios árboles de decisión en secuencia para mejorar la precisión.

###Objetivos de los Modelos de Ensamble
* Mejorar la Precisión del Modelo: Al combinar múltiples modelos, los ensambles capturan patrones complejos en los datos y reducen errores de predicción.

* Reducir la Varianza y el Sesgo: Los modelos de ensamble pueden reducir la varianza (como en el caso de bagging) y el sesgo (como en el boosting), creando modelos que generalizan mejor en nuevos datos.

* Aumentar la Estabilidad: Los modelos de ensamble son menos sensibles a los cambios en los datos, lo cual les permite manejar mejor los datos ruidosos o con variaciones.

````python
import numpy as np
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split

# Separar los datos en conjuntos de entrenamiento y validación
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

# Definir la red neuronal
def create_nn_model():
    model = Sequential()
    model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Entrenar la red neuronal
nn_model = create_nn_model()
nn_model.fit(X_train, y_train, epochs=50, batch_size=10, verbose=1)
nn_preds = nn_model.predict(X_val).flatten()  # Convertir predicciones a una dimensión compatible

# Entrenar Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_preds = rf_model.predict_proba(X_val)[:, 1]  # Usamos la probabilidad de la clase positiva

# Entrenar Gradient Boosting
gb_model = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_model.fit(X_train, y_train)
gb_preds = gb_model.predict_proba(X_val)[:, 1]
````

````python
# Crear el conjunto de características para el modelo de ensamble
stacked_features = np.column_stack((nn_preds, rf_preds, gb_preds))
````

````python
# Realizar predicciones finales con el modelo de ensamble
y_pred_ensemble = stacking_model.predict(stacked_features)

# Calcular y mostrar las métricas de rendimiento
print("Rendimiento del Ensamble (Stacking):")
print("Accuracy:", accuracy_score(y_val, y_pred_ensemble))
print("Precision:", precision_score(y_val, y_pred_ensemble))
print("Recall:", recall_score(y_val, y_pred_ensemble))
print("F1 Score:", f1_score(y_val, y_pred_ensemble))
````

La implementación de un modelo de red neuronal simple consiste en construir una arquitectura de red capaz de aprender patrones complejos en los datos y realizar predicciones precisas. Este tipo de red neuronal, también conocida como red neuronal de una sola capa oculta o perceptrón multicapa, es uno de los modelos fundamentales en el aprendizaje profundo. La red se compone de una capa de entrada, una o más capas ocultas, y una capa de salida. Cada capa de la red transforma los datos a través de neuronas conectadas, ajustando sus pesos para minimizar el error y mejorar la precisión en cada iteración.

Para este caso de estudio, el modelo de red neuronal simple será diseñado para detectar fallos o anomalías, optimizando su capacidad de identificar patrones específicos en los datos. La red será entrenada utilizando un proceso de retropropagación, donde ajustará sus pesos para reducir la diferencia entre las predicciones y los resultados reales. La arquitectura y el proceso de entrenamiento estarán enfocados en maximizar el recall, garantizando que se detecten la mayoría de los fallos, mientras se busca un equilibrio con la precisión para controlar los falsos positivos.

## **Componentes Clave en la Implementación**

**Definición de la Estructura de la Red:**

La red neuronal se construirá con una o más capas ocultas, cada una con un conjunto de neuronas optimizadas para capturar los patrones subyacentes.
La capa de salida usará una función de activación sigmoide, ideal para la clasificación binaria, permitiendo que la red determine la probabilidad de fallo.

**Ajuste de Hiperparámetros:**

Hiperparámetros como el número de neuronas, la tasa de aprendizaje y el número de épocas se definirán de acuerdo con el conjunto de datos para lograr una combinación ideal entre rendimiento y eficiencia.
El ajuste de hiperparámetros también permitirá reducir el sobreajuste, asegurando que la red generalice bien en el conjunto de validación.

**Evaluación y Optimización del Modelo:**

Durante y después del entrenamiento, el modelo será evaluado con métricas clave, como el recall, la precisión, la exactitud y el F1 score, para confirmar su capacidad de detectar fallos y minimizar los falsos positivos.
La evaluación permitirá realizar ajustes adicionales en los hiperparámetros y en la arquitectura de la red, si es necesario.
