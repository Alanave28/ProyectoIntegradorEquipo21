````python
import os
import pandas as pd

# Define the directory path where the Excel files are stored
excel_file_path = '/content/sample_data/'

# Define the expected column names to standardize the structure of the data
column_names = [
    '# Registro', 'Fecha de adquisicin', 'Hora Adquisicin', 'Temperatura Ambiente  (C)',
    'Inverter fault (fault=1)', 'Inverter running', 'Brake contactor feedback (open=1/close=0)',
    'Travelling direction (fwd=0/rev=1)', 'Referencia Velocidad', 'Velocidad Actual',
    'Relative load cell actual weight [%]', 'Delta Temperatura por Velocidad (C) Rod 1',
    'Temperatura Rod 1', 'Delta Temperatura por Velocidad (C) Rod 2', 'Temperatura Rod 2',
    'Delta Temperatura por Velocidad (C) Rod 3', 'Temperatura Rod 3', 'Vibracion punto 1',
    'Vibracion punto 2', 'Vibracion punto 3', 'Motor actual relative current (%)',
    'Motor actual relative current (%).1'
]

# Define a dictionary for renaming specific columns in each file
rename_dict = {
    'Motor actual relative current (%) FWD': 'Motor actual relative current (%)',
    'Motor actual relative current (%) REVERSE': 'Motor actual relative current (%).1'
}

# List to store dataframes created from each Excel file
all_dataframes = []

# Iterate through each file in the specified directory
for filename in os.listdir(excel_file_path):
    if filename.endswith('.xlsx'):
        file_path = os.path.join(excel_file_path, filename)
        try:
            # Read the Excel file into a pandas DataFrame
            df = pd.read_excel(file_path)

            # Rename columns as per the rename_dict to ensure consistency
            df.rename(columns=rename_dict, inplace=True)

            # Add any missing columns from column_names to ensure uniform structure
            for column in column_names:
                if column not in df.columns:
                    df[column] = None  # Fill missing columns with None

            # Reorder columns to match column_names
            df = df[column_names]

            # Append the processed DataFrame to the list
            all_dataframes.append(df)
        except Exception as e:
            print(f"Error reading file {filename}: {e}")

# Concatenate all DataFrames if any were successfully processed
if all_dataframes:
    merged_df = pd.concat(all_dataframes, ignore_index=True)
    print("Files concatenated successfully.")
else:
    print("No matching files with expected columns found in the specified path.")

# Display the first few rows of the merged DataFrame
merged_df.head()
````

En las siguientes funciones vamos a determinar si tenemos datos nulos, contantes y outliers.

````python
import pandas as pd

def check_missing_data(df):
    # Calculate the total number of missing (null) values in each column
    missing_data = df.isnull().sum()

    # Calculate the percentage of missing values for each column
    missing_percentage = (missing_data / len(df)) * 100

    # Create a summary DataFrame with total and percentage of missing values per column
    missing_summary = pd.DataFrame({
        'Total Missing': missing_data,
        'Percentage Missing': missing_percentage
    })

    # Filter the summary to include only columns with missing values
    missing_summary = missing_summary[missing_summary['Total Missing'] > 0]

    # Return the summary sorted by percentage of missing data in descending order
    return missing_summary.sort_values(by='Percentage Missing', ascending=False)

# Usage
missing_summary = check_missing_data(df)
print("Missing Data by Column:\n", missing_summary)

````

````python
def check_constant_columns(df):
    # Identify columns where all values are the same (constant columns)
    constant_columns = [col for col in df.columns if df[col].nunique() == 1]

    # Return the list of constant columns
    return constant_columns

# Usage
constant_columns = check_constant_columns(df)
print("Columns with constant values:\n", constant_columns)

````

````python
def detect_outliers(df, threshold=1.5):
    # Initialize a dictionary to store outlier counts for each column
    outliers = {}

    # Loop through numeric columns (integer and float types only)
    for col in df.select_dtypes(include=['float', 'int']).columns:
        # Calculate the first (Q1) and third quartiles (Q3) and the interquartile range (IQR)
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        # Determine lower and upper bounds for detecting outliers
        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR

        # Count the number of outliers that fall below the lower bound or above the upper bound
        outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col].count()

        # Add to outliers dictionary if any outliers are detected in this column
        if outliers_count > 0:
            outliers[col] = outliers_count

    # Create a summary DataFrame of outliers counts per column
    outliers_summary = pd.DataFrame.from_dict(outliers, orient='index', columns=['Outliers Count'])
    return outliers_summary

# Usage
outliers_summary = detect_outliers(df)
print("Outliers Summary:\n", outliers_summary)

````

````python
# Define limits for each column in a dictionary
limits = {
    'Referencia Velocidad': {'lower': 0, 'upper': 100},
    'Velocidad Actual': {'lower': 0, 'upper': 100},
    'Relative load cell actual weight [%]': {'lower': 0, 'upper': 100},
    'Temperatura Rod 1': {'lower': -10, 'upper': 110},
    'Temperatura Rod 2': {'lower': -10, 'upper': 110},
    'Temperatura Rod 3': {'lower': -10, 'upper': 110},
    'Vibracion punto 1': {'lower': 0, 'upper': 0.15},
    'Vibracion punto 2': {'lower': 0, 'upper': 0.15},
    'Vibracion punto 3': {'lower': 0, 'upper': 0.15}
}

# Function to check if any values exceed specified limits and add a target column 'Y'
def check_limits(df, limits):
    # Initialize a new column 'Y' with 0, assuming all values are within limits
    df['Y'] = 0

    for column, bounds in limits.items():
        lower = bounds.get('lower')
        upper = bounds.get('upper')

        # Update 'Y' to 1 if any value in the column exceeds the lower or upper limit
        df['Y'] = df['Y'] | df[column].apply(lambda x: 1 if (x < lower or x > upper) else 0)

    return df

# Apply the function to the original DataFrame
df = check_limits(final_df, limits)

# Display the first few rows of the modified DataFrame with the 'Y' column
df.head()
````

# **Avance 2. Ingeniera de caractersticas (FE - Feature Engineering)**

El Feature Engineering (Ingeniera de Caractersticas) es el proceso de transformar, seleccionar y crear nuevas variables (features o caractersticas) a partir de los datos originales con el objetivo de mejorar el rendimiento de los modelos de machine learning. Este paso es crucial en el desarrollo de modelos predictivos, ya que las caractersticas bien diseadas y relevantes pueden tener un impacto significativo en la precisin y generalizacin de un modelo.

###Objetivos del Feature Engineering
* Extraer Informacin Relevante: Convertir datos en bruto en caractersticas que mejor representen los patrones del problema.
* Simplificar la Representacin del Problema: Crear caractersticas que hagan el modelo ms simple o ms fcil de interpretar.
* Mejorar la Calidad del Modelo: Generar caractersticas que puedan ayudar al modelo a capturar relaciones complejas y mejorar su capacidad predictiva.

###Principales Tcnicas de Feature Engineering

**Transformacin de Datos:**

* Estandarizacin y Normalizacin: Escalar los datos para que tengan una media y una desviacin estndar uniformes o para que estn en un rango especfico. Esto es til en modelos sensibles a la escala, como redes neuronales y modelos de distancia.
* Transformaciones logartmicas y polinmicas: Aplicar transformaciones como logaritmos o races cuadradas para variables con distribuciones sesgadas o crear nuevas caractersticas cuadrticas/cbicas que capturen relaciones no lineales.

**Manejo de Datos Faltantes:**

* Imputacin: Rellenar valores faltantes con la media, mediana o moda de la columna, o incluso con valores especficos (como un indicador de "desconocido").
* Indicadores de datos faltantes: Crear columnas binarias que indiquen si un valor estaba o no presente, lo cual puede ser informativo.

**Codificacin de Variables Categricas:**

* One-Hot Encoding: Convertir variables categricas en variables binarias (por ejemplo, "rojo", "azul", "verde" se convierten en columnas separadas).
* Target Encoding: Reemplazar valores categricos con el promedio de la variable objetivo (especialmente en problemas de clasificacin o regresin).

**Extraccin de Caractersticas Temporales:**

Para datos con informacin de tiempo, se pueden extraer caractersticas como el da, la semana, el mes, la estacin, si es fin de semana o incluso transformaciones como "hora del da". Estas pueden ser tiles en anlisis de series temporales.

**Interacciones entre Variables:**

Crear caractersticas que representen la interaccin entre dos o ms variables puede ayudar al modelo a capturar relaciones complejas. Por ejemplo, en vez de usar solo el ingreso y la edad de una persona, podras agregar una caracterstica que sea la multiplicacin o el cociente entre ambas.

**Reduccin de Dimensionalidad:**

* PCA (Anlisis de Componentes Principales): Reduce la cantidad de caractersticas en un conjunto de datos mientras se mantiene la mayor parte de la informacin.
* Seleccin de Caractersticas (Feature Selection): Eliminar caractersticas irrelevantes o redundantes, lo cual es til para evitar el sobreajuste y mejorar la eficiencia del modelo.

**UMBRAL DE VARIANZA**

El umbral de varianza es una tcnica de seleccin de caractersticas en el preprocesamiento de datos que ayuda a identificar y eliminar aquellas variables que tienen poca o nula variabilidad en sus valores. Su propsito es reducir la dimensionalidad de los datos al descartar columnas que no contribuyen significativamente al anlisis, ya que las variables con baja varianza suelen contener poca informacin til y no ayudan a distinguir entre distintas clases o patrones en los datos.

**Objetivo del Umbral de Varianza**

El objetivo principal de aplicar un umbral de varianza es optimizar el conjunto de datos para que sea ms eficiente y preciso en las etapas de anlisis y modelado. Las columnas con baja varianza tienden a ser redundantes o constantes, lo cual puede aadir ruido o aumentar la complejidad del modelo sin mejorar su capacidad de generalizacin. Al establecer un umbral mnimo de variabilidad, solo se mantienen las caractersticas que contienen una cantidad significativa de informacin.

**Resultados Esperados**

Al aplicar el umbral de varianza en un conjunto de datos, se esperan los siguientes resultados:

* Reduccin de Dimensionalidad: Eliminacin de columnas constantes o con baja variabilidad, lo cual reduce el nmero de caractersticas y simplifica el conjunto de datos.

* Mejora en la Eficiencia Computacional: Un conjunto de datos con menos columnas es ms rpido de procesar y permite que los algoritmos de aprendizaje automtico entrenen y ejecuten predicciones de manera ms eficiente.

* Aumento en la Precisin del Modelo: Al eliminar caractersticas no informativas, se reduce el ruido en los datos, lo cual puede llevar a una mejora en el rendimiento y la precisin de los modelos predictivos.

* Facilidad de Interpretacin: Con menos variables, el anlisis se vuelve ms manejable y comprensible, permitiendo concentrarse en las caractersticas ms relevantes para el problema de estudio.

````python
from sklearn.feature_selection import VarianceThreshold

# Define the variance threshold
threshold = 0.01  # This is an example; adjust as needed

# Create a feature selector based on the variance threshold
selector = VarianceThreshold(threshold)

# Apply the selector to the DataFrame (excluding the 'Y' column)
X = df.drop(columns=['Y'])
X_selected = selector.fit_transform(X)

# Get the names of columns with variance above the threshold
selected_features = X.columns[selector.get_support(indices=True)]

# Create a new DataFrame with only the selected columns
df_selected = df[selected_features]

# Display the selected columns based on the variance threshold
print("Selected columns based on the variance threshold:")
print(selected_features)

````

**Matriz de Correlacin**

Una matriz de correlacin es una herramienta estadstica que permite medir y visualizar la relacin lineal entre diferentes variables en un conjunto de datos. Cada elemento en la matriz representa el coeficiente de correlacin entre un par de variables, con valores que oscilan entre -1 y 1:

Valores cercanos a 1 indican una fuerte correlacin positiva, lo que significa que, a medida que una variable aumenta, la otra tambin tiende a aumentar.
Valores cercanos a -1 representan una fuerte correlacin negativa, indicando que cuando una variable aumenta, la otra tiende a disminuir.
Valores cercanos a 0 sugieren que no hay una relacin lineal significativa entre las dos variables.


**Objetivo de la Matriz de Correlacin**

El objetivo principal de una matriz de correlacin es identificar relaciones entre variables, lo cual es esencial para el anlisis exploratorio de datos y la seleccin de caractersticas. Esta matriz ayuda a encontrar:

* Patrones y Relaciones Lineales: Permite identificar variables que estn altamente correlacionadas, ya sea de manera positiva o negativa.
* Redundancia: Al encontrar pares de variables con alta correlacin, se pueden eliminar aquellas que son redundantes, reduciendo as la complejidad del modelo.
* Caractersticas Relevantes: Ayuda a seleccionar las variables ms informativas para el anlisis o modelado, mejorando la eficiencia y precisin de los modelos.

**Resultados Esperados**

Al interpretar una matriz de correlacin, los analistas pueden:

* Identificar variables redundantes que pueden eliminarse para simplificar el modelo.
* Seleccionar variables que estn fuertemente correlacionadas con la variable objetivo para mejorar el rendimiento del modelo.
* Detectar relaciones inesperadas que podran requerir un anlisis ms profundo.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create a DataFrame with only the columns selected by the variance threshold
df_var_filtered = df[df_selected.columns]

# Calculate the correlation matrix for the selected columns
correlation_matrix = df_var_filtered.corr().abs()

# Visualize the correlation matrix with a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Selected Columns")
plt.show()

# Identify and remove columns with high correlation (>0.9)
upper_triangle = correlation_matrix.where(
    np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
)

# Find columns with high correlation
high_corr_features = [column for column in upper_triangle.columns if any(upper_triangle[column] > 0.9)]

# Drop highly correlated columns from the DataFrame
df_reduced = df_var_filtered.drop(columns=high_corr_features)

# Display the columns removed due to high correlation
print("Columns removed due to high correlation:")
print(high_corr_features)

# Final result after applying variance threshold and removing high correlation columns
print("Final columns for analysis after removing high correlation:")
print(df_reduced.columns)

````

**PCA**

El Anlisis de Componentes Principales (PCA, por sus siglas en ingls) es una tcnica de reduccin de dimensionalidad utilizada en anlisis de datos y aprendizaje automtico. PCA transforma un conjunto de variables posiblemente correlacionadas en un conjunto ms pequeo de variables no correlacionadas llamadas componentes principales. Estas componentes principales capturan la mayor parte de la variabilidad en los datos originales, permitiendo conservar la informacin relevante mientras se reduce la complejidad del conjunto de datos.

El mtodo se basa en transformar los datos en una nueva base de coordenadas en la que los primeros ejes (componentes) explican la mayor parte de la variacin en los datos. Esto se logra mediante operaciones matemticas como la descomposicin en valores propios de la matriz de covarianza de los datos.

**Objetivos del PCA**

El principal objetivo de PCA es simplificar un conjunto de datos sin perder informacin relevante. Sus objetivos especficos son:

* Reduccin de Dimensionalidad: Al transformar los datos a un espacio de menor dimensin, PCA ayuda a reducir la cantidad de caractersticas, lo cual facilita el procesamiento y anlisis, especialmente cuando el conjunto de datos es amplio.

* Eliminacin de Redundancia: PCA reduce la colinealidad entre variables al encontrar combinaciones lineales de las caractersticas originales, eliminando as la redundancia.

* Mejora de la Interpretabilidad: PCA facilita el anlisis exploratorio al permitir que los datos se representen visualmente en menos dimensiones, como en grficos bidimensionales o tridimensionales que son ms fciles de interpretar.

* Optimizacin en el Desempeo de Modelos: La reduccin de dimensionalidad con PCA puede mejorar la eficiencia y, en algunos casos, la precisin de los modelos de aprendizaje automtico al enfocarse solo en las caractersticas que aportan variabilidad significativa.

* Reduccin del Ruido: PCA puede filtrar el ruido de los datos, ya que las componentes principales suelen captar la variacin significativa, dejando de lado las variaciones que podran ser ruido.

````python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Scale the data before applying PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_reduced.drop(columns=['# Registro']))  # Exclude '# Registro' if it's only an identifier

# Define PCA to retain 95% of the variance
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

# View the amount of variance explained by each principal component
print("Variance explained by each principal component:")
print(pca.explained_variance_ratio_)

# Create a DataFrame with the principal components
pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])

# Display the principal components
print("Principal Components:")
print(pca_df.head())

````

**Support Vector Classifier (SVC)**

El Support Vector Classifier (SVC) es un tipo de modelo de aprendizaje supervisado basado en mquinas de soporte vectorial (Support Vector Machines, SVM) que se utiliza para tareas de clasificacin. SVC busca encontrar un hiperplano ptimo que separe las distintas clases en un conjunto de datos, maximizando la distancia (o margen) entre las clases. Este margen acta como un espacio protector que ayuda a mejorar la precisin del modelo y su capacidad de generalizar a nuevos datos.

SVC es especialmente til cuando los datos no son linealmente separables, ya que puede aplicar el truco del kernel, una tcnica que permite proyectar los datos en un espacio de mayor dimensionalidad donde las clases pueden separarse ms fcilmente. Existen diferentes tipos de kernels (lineal, polinomial, RBF, entre otros) que SVC puede utilizar para adaptar su funcionamiento a los patrones especficos de los datos.

**Objetivos de SVC**

Los principales objetivos del SVC son:

* Maximizar la Separacin entre Clases: Al encontrar el margen ptimo entre clases, SVC mejora la clasificacin y ayuda a reducir errores en la prediccin.

* Flexibilidad en Casos no Lineales: Gracias a los distintos kernels, SVC puede modelar relaciones complejas en los datos, permitiendo una clasificacin efectiva incluso en conjuntos de datos donde las clases no son linealmente separables.

* Generalizacin y Robustez: SVC es conocido por su capacidad de generalizar bien en datos no vistos, ya que el enfoque de maximizar el margen ayuda a evitar el sobreajuste.

**Aplicaciones de SVC**

El SVC se aplica en una variedad de reas como reconocimiento de patrones, clasificacin de imgenes, bioinformtica y deteccin de fraudes, entre otros. Su capacidad para trabajar con datos de alta dimensionalidad y su robustez en tareas de clasificacin complejas lo hacen una opcin preferida para problemas de clasificacin con datos estructurados y desbalanceados.

````python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split

# Separate the features and target variable
# Define 'X' as the feature set with selected columns that could impact the target outcome
# 'y' is the target variable 'Y', assumed to indicate a failure condition
X = df[['Temperatura Ambiente  (C)', 'Referencia Velocidad',
        'Relative load cell actual weight [%]', 'Temperatura Rod 1',
        'Temperatura Rod 2', 'Vibracion punto 1', 'Vibracion punto 2']]
y = df['Y']  # Assuming 'Y' is the target variable

# Split data into training and validation sets
# Use a 70-30 split with a fixed random state for reproducibility
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

# Create the SVM model with an RBF kernel
# The RBF (Radial Basis Function) kernel allows the model to handle non-linear relationships
# 'C' is a regularization parameter (higher values can reduce margin size but may increase overfitting risk)
# 'gamma' controls the influence of each training example (using 'scale' adapts gamma to data distribution)
model = SVC(kernel='rbf', C=1, gamma='scale', random_state=42)  # Adjust 'C' and 'gamma' as needed
model.fit(X_train, y_train)  # Train the SVM model on the training data
````

Las observaciones que requieren pocas divisiones para ser aisladas se consideran anomalas, mientras que aquellas que necesitan ms divisiones se consideran normales. Objetivos de Isolation Forest

* Optimizar el Aislamiento de Outliers: Al construir rboles de aislamiento de forma aleatoria, el modelo busca maximizar el aislamiento de observaciones fuera de lo comn, identificando anomalas con precisin y eficiencia.

* Minimizar el Costo Computacional: Isolation Forest es menos intensivo en clculos que otros algoritmos de deteccin de anomalas, lo que lo hace adecuado para grandes volmenes de datos y entornos de procesamiento en tiempo real.

**K-Means**

Es un algoritmo de agrupamiento (clustering) no supervisado que se utiliza para dividir un conjunto de datos en grupos, o "clusters", basndose en la similitud de los datos. El objetivo de K-Means es organizar los datos en un nmero predefinido de grupos (denotado como K) de forma que los puntos dentro de cada grupo sean lo ms similares entre s y diferentes de los puntos en otros grupos. Este algoritmo es popular en el anlisis exploratorio de datos para descubrir patrones, reducir la dimensionalidad o preparar los datos para otros modelos.

El proceso de K-Means comienza eligiendo aleatoriamente K centros iniciales para los grupos, llamados centroides. Luego, cada punto de datos se asigna al centroide ms cercano, formando as los grupos iniciales. A continuacin, se actualizan los centroides para que representen el centro medio de los puntos en cada grupo. Este proceso de asignacin y actualizacin se repite hasta que los centroides se estabilizan o hasta que se alcanza un nmero mximo de iteraciones.

Objetivos de K-Means

* Agrupar Datos Similares: El objetivo principal es dividir los datos en clusters significativos donde los puntos dentro de cada grupo sean ms similares entre s que a los puntos en otros grupos.

* Minimizar la Distancia Intra-cluster: K-Means busca minimizar la variacin dentro de cada grupo al reducir la distancia de cada punto a su centroide correspondiente, optimizando as la cohesin de los grupos.

* Simplificar el Anlisis de Datos: Al reducir los datos en grupos, K-Means facilita la interpretacin de patrones subyacentes y ayuda a identificar tendencias generales en conjuntos de datos complejos.

````python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

# Scale the data for efficient K-Means clustering
# Scaling ensures that all features contribute equally to the distance calculations in K-Means
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['Temperatura Ambiente  (C)', 'Referencia Velocidad',
        'Relative load cell actual weight [%]', 'Temperatura Rod 1',
        'Temperatura Rod 2', 'Vibracion punto 1', 'Vibracion punto 2']])

# Define and fit the K-Means model
# 'n_clusters=2' assumes 2 clusters: one for normal data and one for potential failures
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_scaled)  # Train the K-Means model on the scaled data

# Assign each data point to a cluster
labels = kmeans.labels_

# Calculate the distance of each point to the centroid of its assigned cluster
# This distance can help identify points close or far from the center, potentially highlighting anomalies
distances = kmeans.transform(X_scaled).min(axis=1)

````

````python
# Define a threshold based on the 95th percentile of the distances
# Points with distances above this threshold will be considered anomalies
threshold = np.percentile(distances, 95)

# Identify anomalous points (potential failures)
# 'anomalies' is a boolean array where 'True' indicates an anomaly
anomalies = distances > threshold
````

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Convert anomalies to binary labels (1 for anomalies/failures, 0 for normal points)
anomaly_labels = anomalies.astype(int)

# Evaluate the performance by comparing with the target variable 'Y'
print("K-Means - Failure Detection Evaluation:")
print("Accuracy:", accuracy_score(df['Y'], anomaly_labels))  # Proportion of correct predictions
print("Precision:", precision_score(df['Y'], anomaly_labels))  # Precision for detecting failures
print("Recall:", recall_score(df['Y'], anomaly_labels))  # Recall for detecting failures
print("F1 Score:", f1_score(df['Y'], anomaly_labels))  # Harmonic mean of precision and recall
````

## **Modelo de Emsamble**
Un modelo de ensamble es una tcnica de aprendizaje automtico que combina mltiples modelos base para crear un modelo ms robusto y preciso. La idea fundamental detrs de los modelos de ensamble es que al combinar modelos individuales, sus errores especficos tienden a cancelarse entre s, produciendo una prediccin final que supera a los modelos individuales en trminos de precisin y estabilidad. Este enfoque se ha convertido en una herramienta popular para resolver problemas complejos en diversas reas, como la clasificacin, la regresin y la deteccin de anomalas.

###Existen principalmente dos tipos de mtodos de ensamble:

**Bagging (Bootstrap Aggregating):**

En este mtodo, cada modelo en el ensamble se entrena en diferentes subconjuntos del conjunto de datos (generalmente mediante muestreo con reemplazo).
Los resultados de todos los modelos se combinan (por promedio en regresin o votacin en clasificacin) para producir la prediccin final.
Ejemplo comn: Random Forest, que utiliza mltiples rboles de decisin para mejorar la estabilidad y reducir el sobreajuste.

**Boosting:**

Boosting entrena modelos de forma secuencial, donde cada modelo intenta corregir los errores de los modelos anteriores.
Los modelos en Boosting suelen tener ms peso en aquellos ejemplos difciles de clasificar.
Ejemplo comn: Gradient Boosting y AdaBoost, que ajustan varios rboles de decisin en secuencia para mejorar la precisin.

###Objetivos de los Modelos de Ensamble
* Mejorar la Precisin del Modelo: Al combinar mltiples modelos, los ensambles capturan patrones complejos en los datos y reducen errores de prediccin.

* Reducir la Varianza y el Sesgo: Los modelos de ensamble pueden reducir la varianza (como en el caso de bagging) y el sesgo (como en el boosting), creando modelos que generalizan mejor en nuevos datos.

* Aumentar la Estabilidad: Los modelos de ensamble son menos sensibles a los cambios en los datos, lo cual les permite manejar mejor los datos ruidosos o con variaciones.
